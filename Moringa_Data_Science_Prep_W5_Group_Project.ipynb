```
df1 = pd.read_csv('Trimed_Flights_data.csv')
df1.head(10)
```

```
# checking our rows and columns
df1.shape
```

```
# checking on columns
df1.columns
```

```
#Change all column names to lower case
df1.columns = map(str.lower, df1.columns)
df1.columns
```

```
# dropping unnecessary columns
df1.drop(['tail_number', 'taxi_out',
       'wheels_off', 'air_time', 
       'wheels_on','taxi_in'],axis= 1, inplace = True)
```

```
#New columns after dropping
df1.columns
```

```
# preview our columns
df1.info()
```

```
# check for null values
df1.isnull().any()
```

```
#After further analysis we saw it was necessary to drop more columns
df1.drop(['cancellation_reason', 'air_system_delay', 'security_delay',
       'airline_delay', 'late_aircraft_delay', 'weather_delay'],axis= 1, inplace = True)
df1.head(5)
```

```
df1.tail()
```

```
# drop null values
df1 = df1.dropna()
df1.info
```

```
#Check for duplicated rows
df1.duplicated()
# and drop if any
df1.drop_duplicates()
#There were no duplicates in this dataset
```

```
# check for unique values
df1.nunique()
```

```
#exporting the clean dataset
df1.to_csv('Aviation_clean.csv')
```
